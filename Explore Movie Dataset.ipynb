```python
base_color = sb.color_palette()[0]
#'popularity','vote_average'
```

```python
#直方图
def draw_hist(key_word, space=1):
    revenue_max = movie_data[key_word].max()
    bins = np.arange(0, revenue_max+space,space)
    plt.hist(data=movie_data, x=key_word, bins=bins, color=base_color);
    plt.xlabel(key_word);
    plt.ylabel('count');

def draw_dist(key_word, space=1):
    revenue_max = movie_data[key_word].max()
    bins = np.arange(0, revenue_max+space,space)
    sb.distplot(movie_data[key_word], bins=bins, kde=True,hist_kws={'alpha' : 0.8});

plt.figure(figsize=(15,6));
plt.subplot(1,2,1)
draw_hist('vote_average', space=0.25);
plt.subplot(1,2,2)
draw_dist('vote_average', space=0.25);
```

（问题回答）通过上图可以看出tmdb-movies.csv 文件中记录的电影的评分vote_average是的分布大致是一个正态分布

**任务8：**（挑战）请尝试挑选一组特征，进行多变量的可视化。多变量的可视化能够帮我们揭示数据之间的关系，例如：电影的票房和知名度的关系。

```python
#电影的评分vote_average 和 电影的欢迎程度popularity的关系 revenue
plt.figure(figsize=(15,5));
plt.subplot(1,2,1)
sb.regplot(movie_data['revenue'],movie_data['vote_average']);
plt.ylabel('vote_average')
plt.xlabel('revenue')
plt.subplot(1,2,2)
#plt.scatter(data=movie_data,x='popularity', y='revenue');
sb.regplot(movie_data['revenue'],movie_data['popularity'])
plt.ylabel('popularity')
plt.xlabel('revenue');
```

```python
def draw_line(key_word, bins=40):
    x_max = movie_data['revenue'].max()
    x_min = movie_data['revenue'].min()
    space = (x_max-x_min)/(bins+1)
    xbin_edges = np.linspace(x_min, x_max, bins+1)
    xbin_centers = (xbin_edges + space/2)[:-1]
    data_xbins = pd.cut(movie_data['revenue'], xbin_edges, right=False, include_lowest=True)
    y_means = movie_data[key_word].groupby(data_xbins).mean()#均值
    y_sems = movie_data[key_word].groupby(data_xbins).sem()#标准误差
    plt.errorbar(x=xbin_centers, y=y_means, yerr=y_sems);
    plt.ylabel(key_word)
```

```python
plt.figure(figsize=(15,5));
plt.subplot(1,2,1)
plt.errorbar(data=movie_data, x='revenue', y='popularity');
plt.ylabel('popularity')
plt.subplot(1,2,2)
draw_line('popularity', bins=30)
```

```python
plt.figure(figsize=(15,5));
plt.subplot(1,2,1)
plt.errorbar(data=movie_data, x='revenue', y='vote_average');
plt.ylabel('vote_average')
plt.subplot(1,2,2)
draw_line('vote_average', bins=30)
```

结论：通过上面研究，可以发现电影的票房(revenue)和知名度(popularity)存在这样的关系：在一定程度上票房越高对应的影片的知名度也相应越高，同样的票房高的影片其评分也不会太差。
